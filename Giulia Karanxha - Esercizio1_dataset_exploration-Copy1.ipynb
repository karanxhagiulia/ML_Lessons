### RELAZIONI CON ALTRE VARIABILI CATEGORICHE

```python
### usare sns.boxplot per plottare ogni variabile categorica (year, fuel, seller_type, transmission, owner) contro il target

for col in ['year', 'fuel', 'seller_type', 'transmission', 'owner', 'torque']:
  plt.figure(figsize=(35,17))
  sns.boxplot(x=col, y="selling_price", data=cars_df)
  plt.xticks(rotation=90, fontsize=13) # ruota le etichette sull'asse x
  plt.show()
```

### CORRELAZIONI

```python
corr_matrix = cars_df.corr()
corr_matrix
corr_matrix['selling_price'].sort_values(ascending=False)
```

```python
sale_corr = corr_matrix['selling_price'].abs().sort_values(ascending=False)

plt.figure(figsize=(10,7))
sns.barplot(x=sale_corr.index, y=sale_corr)
plt.xticks(rotation=90, fontsize=12)
plt.show()
```

```python
sale_corr
```

```python
most_corr_vars = sale_corr.abs().iloc[0:7].index
most_corr_vars
```

```python
 cars_df[most_corr_vars].corr()
```

```python
# plottare la matrice di correlazione con sns.heatmap() e individuare le variabili piÃ¹ correlate con selling_price, oltre che quelle che sembrano essere collineari tra loro
most_corr_vars = sale_corr.abs().iloc[0:7].index
most_corr_matrix = cars_df[most_corr_vars].corr()
#Ã¨ selling price non la quantitÃ  venduta, quindi Ã¨ giusto che il prezzo di vendita Ã¨ inversamente proporzionato al numero di km fatti
plt.figure(figsize=(8,7))
sns.heatmap(most_corr_matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10})
plt.show()
```

Ora vediamo le correlazioni con SalePrice

Volendo Seaborn mette a disposizione la funzione pairplot() per visualizzare gli scatterplot incrociati e le distribuzioni delle variabili...

```python
### plottare il pairplot di seaborn
sns.pairplot(data=cars_df[most_corr_vars]);
```
